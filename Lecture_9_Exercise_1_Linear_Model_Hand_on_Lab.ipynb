<h3>1. Hồi quy tuyến tính đơn biến và hồi quy tuyến tính đa biến</h3>


<h4>Hồi quy tuyến tính đơn biến</h4>


<p>Một ví dụ về Mô hình dữ liệu mà chúng ta sẽ sử dụng là</p>
<b>Hồi quy tuyến tính đơn biến</b>.

<br>
<p>Hồi quy tuyến tính đơn biến là phương pháp giúp chúng ta hiểu được mối quan hệ giữa hai biến:</p>
<ul>
    <li>Biến dự báo/biến độc lập (X)</li>
    <li>Biến mục tiêu/biến phụ thuộc (mà chúng ta muốn dự đoán) (Y)</li>
</ul>

<p>Kết quả của Hồi quy tuyến tính là một <b>hàm tuyến tính</b> dự đoán biến mục tiêu (phụ thuộc) dưới dạng hàm của biến dư báo (độc lập)
</p>


 <b>Hàm tuyến tính:</b>
$$
Yhat = a + b  X
$$


<ul>
    <li>a: Điểm cắt của đường thẳng hồi quy và trục Y (intercept)</li>
    <li>b: Hệ số góc (slope) </li>
</ul>

<p>Nếu muốn sử dụng nhiều biến hơn trong mô hình của mình để dự đoán giá ô tô, chúng ta có thể sử dụng <b>Hồi quy tuyến tính đa biến</b>.
    


$$
Y:Biến\ mục \ tiêu \ 1 \ \ \\
X_1:Biến\ dự \ đoán \ 1 \ \ \\
X_2:Biến\ dự \ đoán \ 2 \ \ \\
X_3:Biến\ dự \ đoán \ 3 \ \ \\

$$
a: intercept\\ \ \  
b_1:Hệ \ số \ của \ biến \ 1 \ \  \\

b_4:Hệ \ số \ của \ biến \ 4 \ \  \\
$$

$$
Yhat = a + b_1 X_1 + b_2 X_2 + b_3 X_3 + b_4 X_4

<h3>2) Đánh giá mô hình sử dụng công cụ trực quan</h3>


Bây giờ chúng ta đã phát triển một số mô hình, làm cách nào để đánh giá các mô hình của mình và làm cách nào để chọn mô hình tốt nhất? Một cách để làm điều này là sử dụng công cụ trực quan.

Import thư viện trực quan: seaborn

```python
import seaborn as sns
%matplotlib inline
```

<h3>Regression Plot</h3>


<p>Khi nói đến hồi quy tuyến tính đơn biến, một cách tuyệt vời để trực quan hóa sự phù hợp của mô hình của chúng ta là sử dụng <b>Regression plots</b>.</p>

<p>Biểu đồ này hiển thị sự kết hợp giữa <b>biểu đồ phân tán</b>, cũng như biểu đồ <b>hồi quy tuyến tính</b>. Điều này sẽ cho chúng ta ước tính sự hợp lý về mối quan hệ giữa hai biến, mức độ tương quan (âm, dương).</p>

Vẽ biểu đồ regession plot giữa 2 biến `highway-mpg` và biến `price` 

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.regplot(x="highway-L/100km", y="price", data=df)
plt.ylim(0,)
```

Từ biểu đồ này, chúng ta có thể thấy rằng biến `price` có mối tương quan thuận với `highway-L/100km`, vì `slope` của đường hồi quy là dương. Một điều cần lưu ý khi xem biểu đồ hồi quy là chú ý đến mức độ phân tán của các điểm dữ liệu xung quanh đường hồi quy. Nếu dữ liệu ở quá xa đường thẳng hồi quy, mô hình tuyến tính này có thể không phải là mô hình tốt nhất cho tập dữ liệu này. 


Vẽ biểu đồ regession plot giữa 2 biến `peak-rpm` và biến `price`

```python
plt.figure(figsize=(width, height))
sns.regplot(x="peak-rpm", y="price", data=df)
plt.ylim(0,)
```

So sánh biểu đồ hồi quy của peak-rpm" và "highway-L/100km", chúng ta có nhận thấy các điểm dữ liệu của biến "highway-L/100km" nằm gần hơn với đường hồi quy và có xu hướng tương quan dương với mức giá. Ngược lại, các điểm dữ liệu của biến "peak-rpm" phân tán  ngẫu nhiên so với với đường hồi quy, và rất khó xác định xu hướng của các điểm này tăng hay giảm khi mà biến "peak-rpm" tăng

<h3>Residual Plot</h3>

<p>Một cách tốt để hình dung phương sai của dữ liệu là sử dụng biểu sai số</p>

<p>Biểu đồ <b>residual plot</b>?</p>

<p>Biểu đồ <b>residual plot</b> mô tả sự khác biệt giữa giá trị quan sát được (y) và giá trị dự đoán (Yhat) được gọi là phần dư (e). Khi chúng ta xem biểu đồ hồi quy, phần dư là khoảng cách từ điểm dữ liệu đến đường hồi quy.</p>

<p>Vậy biểu đồ <b>residual plot</b> là gì?</p>

<p>Residuals Plot cho thấy sự khác biệt giữa phần dư trên trục tung và giá trị dự đoán ở trục hoành</p>

<p>Khi xét biểu đồ Residuals Plot chúng ta cần chú ý điều gì?</p>

<p>Chúng ta cần xem xét sự phân tán của phần dư:</p>

<p>- Nếu các điểm trong residual plot <b>được phân tán ngẫu nhiên xung quanh trục x</b> thì <b>mô hình tuyến tính được coi là phù hợp</b> cho dữ liệu. Tại sao vậy? Phần dư phân tán ngẫu nhiên nó nghĩa là phương sai không đổi và do đó mô hình tuyến tính rất phù hợp với dữ liệu này.</p>


```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['highway-L/100km'], y=df['price'])
plt.show()
```
